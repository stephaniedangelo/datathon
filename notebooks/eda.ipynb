```python
# Carregar o JSON puro
import json

with open('../data/prospects.json', 'r', encoding='utf-8') as f:
    raw_prospects = json.load(f)
```

```python
# Lista para armazenar os dados extraídos
rows = []

# Percorrer cada vaga (uma vaga = uma linha do JSON)
for vaga_id, prospect in raw_prospects.items():
    candidatos = prospect.get("prospects", [])
    for candidato in candidatos:
        codigo = candidato.get("codigo")
        nome = candidato.get("nome")
        status = candidato.get("status", "").lower()  # pode ser 'contratado', 'reprovado', etc.
        contratado = 1 if "contratado" in status else 0
        rows.append({
            "codigo_vaga": vaga_id,
            "codigo_profissional": codigo,
            "nome": nome,
            "contratado": contratado
        })
```

```python
prospects_df = pd.DataFrame(rows)
prospects_df.head()
```

```python
prospects_df['codigo_profissional'] = prospects_df['codigo_profissional'].astype(str)
```

```python
prospects_df['codigo_vaga'] = prospects_df['codigo_vaga'].astype(str)
```

```python
print(applicants_df.columns)
print(vagas_df.columns)
```

```python
vagas_df = vagas_df.reset_index().rename(columns={'index': 'codigo_vaga'})
vagas_df['codigo_vaga'] = vagas_df['codigo_vaga'].astype(str)
```

```python
applicants_df['codigo_profissional'] = applicants_df['codigo_profissional'].astype(str)
prospects_df['codigo_profissional'] = prospects_df['codigo_profissional'].astype(str)
prospects_df['codigo_vaga'] = prospects_df['codigo_vaga'].astype(str)
```

```python
# Junta dados do candidato
df = pd.merge(prospects_df, applicants_df, on='codigo_profissional', how='left')

# Junta dados da vaga
df = pd.merge(df, vagas_df, on='codigo_vaga', how='left')
```
